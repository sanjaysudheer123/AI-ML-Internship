```python
df['Age_mean_imputed']=df['Age'].fillna(df['Age'].mean())
df['Age_median_imputed']=df['Age'].fillna(df['Age'].median())
```

```python
df['Cabin_mode_imputed']=df['Cabin'].fillna(df['Cabin'].mode()[0])
```

```python
#create plotstyle
sns.set(style="whitegrid")

#create boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df,x='Embarked',y='Fare',palette='pastel')

#set plot title and labels
plt.title('Box Plot of Fare by Embarked',fontsize=12)
plt.xlabel('Embarked',fontsize=12)
plt.ylabel('Fare',fontsize=12)

#show plot
plt.show()
```

```python
encoded_data=df.copy()
```

```python
#Label Encoder for Sex

label_encoder=LabelEncoder()
encoded_data['Sex_Label']=label_encoder.fit_transform(encoded_data['Sex'])
```

```python
one_hot_encoded_data=pd.get_dummies(encoded_data['Embarked'],prefix='Embarked')
encoded_data=pd.concat([encoded_data,one_hot_encoded_data],axis=1)
```
