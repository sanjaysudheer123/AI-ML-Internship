```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
df=pd.read_csv('/content/nlp_dataset.csv')
df
```

```python
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab')
```

```python
stop_words=set(stopwords.words('english'))
print(stop_words)
lematizer=WordNetLemmatizer()
```

```python
def process_text(text):
  word_tokens=word_tokenize(text)
  filtered_tokens=[word for word in word_tokens if word.lower() not in stop_words] # Corrected 'woed' to 'word' and 'token' to 'word'
  lemmatized_tokens=[lematizer.lemmatize(word.lower()) for word in filtered_tokens] # Added '()' to word.lower
  return{
      'original_text':text,
      'word_tokens':word_tokens,
      'filtered_tokens':filtered_tokens,
      'lemmatized_tokens':lemmatized_tokens
  }
```

```python
df['Processed_Text'] = df['Text'].apply(process_text)
df.head()
```

```python
results=df["Text"].apply(process_text)


process_df=pd.DataFrame(results.tolist())

print(process_df.head())

processsed_file_path="processed_nlp_dataset.csv"
process_df.to_csv(processsed_file_path,index=False)
print(f"Processed data saved to {processsed_file_path}")
```
